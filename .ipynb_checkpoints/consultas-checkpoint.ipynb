```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
comunas = pd.read_csv("data/comunas.csv")
plebiscitoAR = pd.read_csv("data/Plebiscito Apruebo-Rechazo.csv", sep=";", decimal=",")

plebiscitoC = pd.read_csv("data/Plebiscito convención.csv", sep=";", decimal=",")
presidencial1ra = pd.read_csv("data/resultadosPresidencial1era.csv")
presidencial2da = pd.read_csv("data/resultadosPresidencial2da.csv")

ingresos = pd.read_csv("data/porcentajePobreza.csv")
candidatos1era = list(presidencial1ra.columns)[1:]
candidatos2da = list(presidencial2da.columns)[1:]
piniera1era = candidatos1era.index('Sebastian Piniera Echenique')
piniera2da = candidatos2da.index('Sebastian Piniera Echenique')
```

```python
plebiscito_convencion = plebiscitoC[['Comuna', 'Convencion Mixta (1)', 'Convencion Constitucional (0)']].set_index('Comuna').rename(columns={"Convencion Mixta (1)": "Convencion Mixta", "Convencion Constitucional (0)": "Convencion Constitucional"})
plebiscito_apruebo_rechazo = plebiscitoAR[['Comuna', 'Apruebo (1)', 'Rechazo (0)']].set_index('Comuna').rename(columns={"Apruebo (1)": "Apruebo", "Rechazo (0)": "Rechazo"})
plebiscito = plebiscito_apruebo_rechazo.join(plebiscito_convencion)
plebiscito
```

## Cálculo del peso de votar por piñera en el apruebo

```python
general = plebiscitoAR.join(ingresos.set_index("comuna"), on = "Comuna")
general = general.join(presidencial2da.set_index("comuna"), on="Comuna")
general.head()
```

```python
## probabilidad categoria ingreso

values = general["categoria"].value_counts()
categ = {}
for i in values.index:
    categ[i] = values[i]/comunas.shape[0]
categ
```

```python
values.index
```

```python
general["Sebastian Piniera Echenique"] > general["Alejandro Guillier Alvarez"]
```

```python
general["ganador"] == piniera2da
```

```python
## probabilidad apruebo que voto por piñera 
init = datetime.now()
apruebo = 1
aprobPiniera = 0
for i in values.index:
    match = general.loc[(general["Apruebo (1)"] > general["Rechazo (0)"]) & (general["Sebastian Piniera Echenique"] > general["Alejandro Guillier Alvarez"]) & (general["categoria"]==i)]
    total = general.loc[(general["ganador"]==piniera2da) & (general["categoria"]==i)]
    if(not total.shape[0]):
        continue
    aprobPiniera += (match.shape[0]/total.shape[0]) * categ[i]
    
aprobNoPiniera = 0
## probabilidad apruebo que no voto por piñera
for i in values.index:
    match = general.loc[(general["Ganador"]==apruebo) & (general["ganador"]!=piniera2da) & (general["categoria"]==i)]
    total = general.loc[(general["ganador"]!=piniera2da) & (general["categoria"]==i)]
    if(not total.shape[0]):
        continue
    aprobNoPiniera += (match.shape[0]/total.shape[0]) * categ[i]

## probabilidad rechazo que voto por piñera 
rechazo = 0
rechazoPiniera = 0
for i in values.index:
    match = general.loc[(general["Ganador"]==rechazo) & (general["ganador"]==piniera2da) & (general["categoria"]==i)]
    total = general.loc[(general["ganador"]==piniera2da) & (general["categoria"]==i)]
    if(not total.shape[0]):
        continue
    rechazoPiniera += (match.shape[0]/total.shape[0]) * categ[i]

## probabilidad rechazo que no voto por piñera 
rechazoNoPiniera = 0
for i in values.index:
    match = general.loc[(general["Ganador"]==rechazo) & (general["ganador"]!=piniera2da) & (general["categoria"]==i)]
    total = general.loc[(general["ganador"]!=piniera2da) & (general["categoria"]==i)]
    if(not total.shape[0]):
        continue
    rechazoNoPiniera += (match.shape[0]/total.shape[0]) * categ[i]

meanPiniera = 1*aprobPiniera + 0*rechazoPiniera
meanNoPiniera = 1*aprobNoPiniera + 0*rechazoNoPiniera
print(f"Peso voto por Piñera: {meanPiniera - meanNoPiniera}")
meanPiniera = 1*aprobPiniera - 1*rechazoPiniera
meanNoPiniera = 1*aprobNoPiniera - 1*rechazoNoPiniera
print(f"Peso voto por Piñera (con -1): {meanPiniera - meanNoPiniera}")
final = datetime.now()
print(f"Tiempo de ejecución: {final-init}")
```

```python

init = datetime.now()
ganador2Cond1 = {}
for i in ganador2:
    match = general.loc[(general["ganador2"]==i) & (general["ganador"]==piniera1era)]
    total = general.loc[general["ganador"]==piniera1era]
    if(not total.shape[0]):
        continue
    ganador2Cond1[i] = match.shape[0]/total.shape[0]


apruebo = 1
aprobPiniera = 0
for i in values.index:
    for j in ganador2:
        match = general.loc[(general["Ganador"]==apruebo) & (general["ganador"]==piniera1era) & (general["categoria"]==i) & (general["ganador2"]==j)]
        total = general.loc[(general["ganador"]==piniera1era) & (general["categoria"]==i) & (general["ganador2"]==j)]
        if(not total.shape[0]):
            continue
        aprobPiniera += (match.shape[0]/total.shape[0]) * categ[i] * ganador2[j] * ganador2Cond1[j]

aprobNoPiniera = 0
## probabilidad apruebo que no voto por piñera
for i in values.index:
    for j in ganador2:
        match = general.loc[(general["Ganador"]==apruebo) & (general["ganador"]!=piniera1era) & (general["categoria"]==i) & (general["ganador2"]==j)]
        total = general.loc[(general["ganador"]!=piniera1era) & (general["categoria"]==i) & (general["ganador2"]==j)]
        if(not total.shape[0]):
            continue
        aprobNoPiniera += (match.shape[0]/total.shape[0]) * categ[i] * ganador2[j] * ganador2Cond1[j]


## probabilidad rechazo que voto por piñera 
apruebo = 0
rechazoPiniera = 0
for i in values.index:
    for j in ganador2:
        match = general.loc[(general["Ganador"]==apruebo) & (general["ganador"]==piniera1era) & (general["categoria"]==i) & (general["ganador2"]==j)]
        total = general.loc[(general["ganador"]==piniera1era) & (general["categoria"]==i) & (general["ganador2"]==j)]
        if(not total.shape[0]):
            continue
        rechazoPiniera += (match.shape[0]/total.shape[0]) * categ[i] * ganador2[j] * ganador2Cond1[j]


## probabilidad rechazo que no voto por piñera 
apruebo = 0
rechazoNoPiniera = 0
for i in values.index:
    for j in ganador2:
        match = general.loc[(general["Ganador"]==apruebo) & (general["ganador"]!=piniera1era) & (general["categoria"]==i) & (general["ganador2"]==j)]
        total = general.loc[(general["ganador"]!=piniera1era) & (general["categoria"]==i) & (general["ganador2"]==j)]
        if(not total.shape[0]):
            continue
        rechazoPiniera += (match.shape[0]/total.shape[0]) * categ[i] * ganador2[j] * ganador2Cond1[j]

meanPiniera = 1*aprobPiniera + 0*rechazoPiniera
meanNoPiniera = 1*aprobNoPiniera + 0*rechazoNoPiniera
print(f"Peso voto por Piñera: {meanPiniera - meanNoPiniera}")
meanPiniera = 1*aprobPiniera - 1*rechazoPiniera
meanNoPiniera = 1*aprobNoPiniera - 1*rechazoNoPiniera
print(f"Peso voto por Piñera (con -1): {meanPiniera - meanNoPiniera}")
final = datetime.now()
print(f"Tiempo de ejecución: {final-init}")
```

```python
def count_by_poor_category(category=0):
    ## probabilidad apruebo que voto por piñera en la segunda vuelta
    apruebo = 1
    aprobPiniera = 0
    match = general.loc[(general["Ganador"]==apruebo) & (general["ganador"]==piniera2da) & (general["categoria"]==category)]
    total = general.loc[(general["ganador"]==piniera2da) & (general["categoria"]==category)]
    if(total.shape[0]):
        aprobPiniera += (match.shape[0]/total.shape[0]) * categ[category]

    aprobNoPiniera = 0
    ## probabilidad apruebo que no voto por piñera en la segunda vuelta
    match = general.loc[(general["Ganador"]==apruebo) & (general["ganador"]!=piniera2da) & (general["categoria"]==category)]
    total = general.loc[(general["ganador"]!=piniera2da) & (general["categoria"]==category)]
    if(total.shape[0]):
        aprobNoPiniera += (match.shape[0]/total.shape[0]) * categ[category]

    ## probabilidad rechazo que voto por piñera 
    apruebo = 0
    rechazoPiniera = 0
    match = general.loc[(general["Ganador"]==apruebo) & (general["ganador"]==piniera2da) & (general["categoria"]==category)]
    total = general.loc[(general["ganador"]==piniera2da) & (general["categoria"]==category)]
    if(total.shape[0]):
        rechazoPiniera += (match.shape[0]/total.shape[0]) * categ[category]

    ## probabilidad rechazo que no voto por piñera 
    apruebo = 0
    rechazoNoPiniera = 0
    match = general.loc[(general["Ganador"]==apruebo) & (general["ganador"]!=piniera2da) & (general["categoria"]==category)]
    total = general.loc[(general["ganador"]!=piniera2da) & (general["categoria"]==category)]
    if(total.shape[0]):
        rechazoNoPiniera += (match.shape[0]/total.shape[0]) * categ[category]

    meanPiniera = 1*aprobPiniera + 0*rechazoPiniera
    meanNoPiniera = 1*aprobNoPiniera + 0*rechazoNoPiniera
    print(f"Peso voto por Piñera: {meanPiniera - meanNoPiniera}")
    meanPiniera = 1*aprobPiniera - 1*rechazoPiniera
    meanNoPiniera = 1*aprobNoPiniera - 1*rechazoNoPiniera
    print(f"Peso voto por Piñera (con -1): {meanPiniera - meanNoPiniera}")
    

```

## Construcción de nuevo Dataset con elecciones presidenciales anteriores

Como puede verse, hay un efecto positivo de votar por Piñera en las elecciones solo si la comuna está en el porcentaje de pobreeza más bajo. Esta tendencia histórica puede verse en cada elección porque comunas con excelente situación económica, como Lo Barnechea, La Dehesa o Las Condes, frecuentemente tienen voto mayoritario de derecha.

```python
segunda_vuelta = pd.read_csv("data/Elecciones segunda vuelta.csv", sep=";", decimal=",", index_col='Comuna')
```

```python
pobreza = general.set_index('Comuna')['categoria']
pobreza.rename(columns={'categoria': 'Decil_pobreza'})
```

```python
info = segunda_vuelta.join(plebiscito).join(pobreza).rename(columns={'categoria': 'Decil_pobreza'})
info = info.fillna(0)

derecha_2000 = info['Joaquin Lavin']
izquierda_2000 = info['Ricardo Lagos']

derecha_2006 = info['Sebastian Piñera']
izquierda_2006 = info['Michelle Bachelet']

derecha_2010 = info['Sebastian Piñera.1']
izquierda_2010 = info['Eduardo Frei']

derecha_2013 = info['Evelyn Matthei Fornet']
izquierda_2013 = info['Michelle Bachelet Jeria']

izquierda_2017 = info['Alejandro Guillier Alvarez']
derecha_2017 = info['Sebastian Piñera Echenique']

apruebo = info['Apruebo']
rechazo = info['Rechazo']

gana_derecha_2000 = derecha_2000 > izquierda_2000
gana_izquierda_2000 = izquierda_2000 > derecha_2000

gana_derecha_2006 = derecha_2006 > izquierda_2006
gana_izquierda_2006 = izquierda_2006 > derecha_2006

gana_derecha_2010 = derecha_2010 > izquierda_2010
gana_izquierda_2010 = izquierda_2010 > derecha_2010

gana_derecha_2013 = derecha_2013 > izquierda_2013
gana_izquierda_2013 = izquierda_2013 > derecha_2013

gana_derecha_2017 = derecha_2017 > izquierda_2017
gana_izquierda_2017 = izquierda_2017 > derecha_2017

gana_apruebo = apruebo > rechazo
gana_rechazo = rechazo > apruebo

info
```

```python
values = info["Decil_pobreza"].value_counts()
values
```

```python
influencia_en_apruebo_derecha_2000 = 0
influencia_en_apruebo_izquierda_2000 = 0
influencia_en_rechazo_derecha_2000 = 0
influencia_en_rechazo_izquierda_2000 = 0

for i in values.index:
    match = info.loc[(gana_apruebo) & (gana_derecha_2000) & (info["Decil_pobreza"]==i)]
    total = info.loc[(gana_derecha_2000) & (info["Decil_pobreza"]==i)]
    if(total.shape[0]):
        influencia_en_apruebo_derecha_2000 += (match.shape[0]/total.shape[0]) * categ[i]

    match = info.loc[(gana_apruebo) & (gana_izquierda_2000) & (info["Decil_pobreza"]==i)]
    total = info.loc[(gana_izquierda_2000) & (info["Decil_pobreza"]==i)]
    if(total.shape[0]):
        influencia_en_apruebo_izquierda_2000 += (match.shape[0]/total.shape[0]) * categ[i]

    match = info.loc[(gana_rechazo) & (gana_derecha_2000) & (info["Decil_pobreza"]==i)]
    total = info.loc[(gana_derecha_2000) & (info["Decil_pobreza"]==i)]
    if(total.shape[0]):
        influencia_en_rechazo_derecha_2000 += (match.shape[0]/total.shape[0]) * categ[i]

    match = info.loc[(gana_rechazo) & (gana_izquierda_2000) & (info["Decil_pobreza"]==i)]
    total = info.loc[(gana_izquierda_2000) & (info["Decil_pobreza"]==i)]
    if(total.shape[0]):
        influencia_en_rechazo_izquierda_2000 += (match.shape[0]/total.shape[0]) * categ[i]


influencia_derecha_2000 = 1*influencia_en_apruebo_derecha_2000 + 0*influencia_en_rechazo_derecha_2000
influencia_izquierda_2000 = 1*influencia_en_apruebo_izquierda_2000 + 0*influencia_en_rechazo_izquierda_2000

print('Influencia voto por derecha 2000 en apruebo/rechazo: ', influencia_derecha_2000 - influencia_izquierda_2000)
print('Influencia voto por izquierda 2000 en apruebo/rechazo: ', influencia_izquierda_2000 - influencia_derecha_2000)

print()

influencia_derecha_2000 = influencia_en_apruebo_derecha_2000 - influencia_en_rechazo_derecha_2000
influencia_izquierda_2000 = influencia_en_apruebo_izquierda_2000 - influencia_en_rechazo_izquierda_2000

print('Influencia voto por derecha 2000 en apruebo/rechazo: ', influencia_derecha_2000 - influencia_izquierda_2000)
print('Influencia voto por izquierda 2000 en apruebo/rechazo: ', influencia_izquierda_2000 - influencia_derecha_2000)
```
